```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data_path = "./data/Bike_rental_modified.csv"
header = [
    "instant", 
    "dteday",#ordinal
    "season",#ordinal
    "yr",#ordinal
    "mnth",#ordinal
    "holiday", #ordinal
    "weekday",#ordinal
    "workingday",#ordinal
    "weathersit",#ordinal
    "temp", #metrisch
    "atemp",#metrisch
    "hum",#metrisch
    "windspeed",#metrisch
    "casual",#metrisch
    "registered",#metrisch
    "cnt",#metrisch
    "intday"
]
df = pd.read_csv(data_path)
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(1,figsize=(25,5))
ax.plot(df["instant"], df["cnt"], label="count of rented bikes", color = "purple")
ax.set_xlabel("Tage seit dem 1.1.2011")
ax.legend()
```

```python
fig, ax = plt.subplots(1,figsize=(25,5))
ax.plot(df["instant"], df["cnt"], label="count of rented bikes", color = "purple")
ax.plot(df["instant"], df["registered"], label="count of registered rented bikes", color = "green")

ax.plot(df["instant"], df["casual"], label="count of casual rented bikes", color = "red")

ax.set_xlabel("Tage seit dem 1.1.2011")
ax.legend()
```

```python
for col in header:
    fig, ax = plt.subplots(1,figsize=(25,5))
    ax.plot(df["instant"], df[col], label=col, color = "purple")
    ax.set_xlabel("Tage seit dem 1.1.2011")
    ax.legend()
```

```python
n = len(df.columns)

fig, ax = plt.subplots(n, figsize=(25,2*n))

i= 0
for key in df.columns:
    ax[i].plot(df["instant"], df[key], label = key)
    ax[i].legend()
    i = i+1 
```
